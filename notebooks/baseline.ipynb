```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_palette("husl")
```

```python
log_path = "../logs/baseline_training_log.csv"
log_df = pd.read_csv(log_path)
log_df.head()
```

```python
plt.plot(log_df["episode"], log_df["loss"])
plt.xlabel("Episode")
plt.ylabel("Loss")
plt.title("Training Loss - JAX Cartpole Baseline")
plt.show()
```
